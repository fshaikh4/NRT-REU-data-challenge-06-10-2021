```python
grouded_data = data_filtered.groupby(["State", "Ten-Year Age Groups Code"]).agg({"Deaths": "sum"})#.sort_values("Deaths", axis = 0, ascending = False)
# grouded_data
grouded_data["total_population"] = data_filtered.groupby(["State", "Ten-Year Age Groups Code"]).agg({"Population": "sum"})
grouded_data["proportion"] = grouded_data["Deaths"] /grouded_data["total_population"] * 100

grouded_data.sort_values("proportion", axis = 0, ascending = False).head(50)
```
